```python
import os
import pandas as pd

import warnings

warnings.filterwarnings("ignore")

strategies = os.listdir("data/strategies")

total_unk = 0
total_short = 0
total_long = 0
total = 0

for ticker in strategies:
    if "-" not in ticker:
        df = pd.read_parquet(f"data/strategies/{ticker}")
        df = df[df.posicao.isnull() == False]
        total += len(df)

        total_unk += len(df[df.posicao == "UNKNOWN"])
        total_short += len(df[df.posicao == "SHORT"])
        total_long += len(df[df.posicao == "LONG"])

short_porc = total_short / total
long_porc = total_long / total
unk_porc = total_unk / total

print(f"unk: {unk_porc}" f" short: {short_porc}" f" long: {long_porc}")
short_porc + long_porc + unk_porc
```
